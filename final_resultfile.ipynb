```python
#

#
import os
#
#
# 

#
import hashlib

def calculate_hash(file_path):
    hash_object = hashlib.sha256()
    with open(file_path, "rb") as file:
        while True:
            data = file.read(65536)  # 
            if not data:
                break
            hash_object.update(data)
    return hash_object.hexdigest()


#
def delete_files(folder_path,want_to_delete_name):
    for delete_name in want_to_delete_name:
        file_to_delete = f"{folder_path}{delete_name}"
        if os.path.exists(file_to_delete):
            os.remove(file_to_delete)
            print(f"{file_to_delete} ")
        else:
            print(f"{file_to_delete} ")

def get_filelist ():
    folder_path = f'name_breaker/'
    file_list = os.listdir(folder_path)

    # 
    file_list = [file_name for file_name in file_list if file_name.endswith('.properties')]
    #print(file_list)
    file_compare(folder_path,file_list)
            
# 
def file_compare(folder_path,file_list):
    want_to_delete_name = []
    
    now_topcount = 0
    for file_name in file_list:
        delete_file_flag = False
        with open(os.path.join(folder_path, file_name), 'r') as file:
            lines = file.readlines()
        # 
        for other_file_name in file_list[now_topcount:]:
            if file_name == other_file_name:
                continue  # 

            hash_A = calculate_hash(f"{folder_path}{file_name}")
            hash_B = calculate_hash(f"{folder_path}{other_file_name}")

            if hash_A == hash_B:
                #print("AB")
                delete_file_flag =True
                #print(f"{file_name} : {other_file_name}")
                want_to_delete_name.append(other_file_name)

            """with open(os.path.join(folder_path, other_file_name), 'r') as other_file:
                other_lines = other_file.readlines()

            if lines == other_lines:
                delete_file_flag =True
                #print(f"{file_name} : {other_file_name}")
                want_to_delete_name.append(other_file_name)"""
        #
        now_topcount += 1
        if delete_file_flag:
            print(f"same file with {file_name} are delete {now_topcount}")
            now_topcount -= len(want_to_delete_name)
            delete_files(folder_path,want_to_delete_name)
            break
    return delete_file_flag
```

```python
#

import difflib
import glob

def get_file_data(filename1,filename2):
    #filenames = glob.glob(f"name_breaker/" + "*")

    with open(filename1, "r") as properties_file:
        a = properties_file.readlines()
    with open(filename2, "r") as properties_file:
        b = properties_file.readlines()

    # Differ
    differ = difflib.Differ()

    # 
    compare_result = differ.compare(a,b)
    
    return compare_result

def make_similler_per(compare_result):
    #
    #print(f"before : {file_before}")
    #print(f"after : {file_after}")
    # 
    #print(result)
    pluse = 0
    mina = 0
    ques = 0
    length = 0
    for i in compare_result:
        #print(i)
        length += 1
        if(i.split()[0] == "+"):
            pluse += 1
        elif (i.split()[0] == "-"):
            mina += 1
        elif (i.split()[0] == "?"):
            ques += 1
            
    return pluse,mina,ques,length

def sim_reu(each_keys,sim_percent,i,j):
    pluse = each_keys[0]
    mina = each_keys[1]
    ques = each_keys[2]
    length = each_keys[3]
    #print(f"{pluse} : {pluse/length}  ,,, {mina} : {mina/length}")
    #print(f"{(length - pluse - mina - ques)/(length - ques)}")
    rs = round((length - pluse - mina - ques)/(length - ques),5)
    sim_percent.append([rs,i,j])
    return sim_percent
        
def eatfile(pluse,big_filename,small_filename,eat_result):
    if pluse == 0 and small_filename in eat_result:
        
        #
        added_part = eat_result[small_filename]
        added_part.append(big_filename)
        
        eat_result[small_filename] = added_part
    elif pluse == 0:
        print(f"small {small_filename} : extend {big_filename}")
        eat_result[small_filename] = [big_filename]
    return eat_result

#
small_eat_result = {}
sim_percent = []

filenames = glob.glob(f"name_breaker/" + "*")
for i in filenames:
    for j in filenames:
        if i == j:
            continue
        compare_result = get_file_data(i,j)
        each_keys = make_similler_per(compare_result)
        
        sim_percent = sim_reu(each_keys,sim_percent,i,j)
        small_eat_result = eatfile(each_keys[0],i,j,small_eat_result)
```
