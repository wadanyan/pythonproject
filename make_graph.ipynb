```python
def step1(file_cansee_number):
    #一つ済みファイル内の要素(user_keyごと)を見ていって、同じのがなければ追加する
    valu = list(file_cansee_number.values())

    connect_val = []

    for i in valu:
        #resu_connect_val = []
        for one_list in i:

            check_exsist_in_list = list(set(one_list))
            if not check_exsist_in_list in connect_val:
                connect_val.append(check_exsist_in_list)

        #connect_val.extend(resu_connect_val)
    connect_val.sort(key=len)
    return  connect_val
```

```python
def create_file_num_list_unique(file_cansee_number):
    
    connect_val = []
    valu = list(file_cansee_number.values())
    for i in valu:
        for one_list in i:
        
            check_exsist_in_list = list(set(one_list))
            if not check_exsist_in_list in connect_val:
                connect_val.append(check_exsist_in_list)

    return connect_val
    
def find_diff_num_list(ListA,connect_val):

    for ListB in connect_val:
        #listC = set(ListA) ^ set(ListB)
        #listC = list(listC)
        #listC = set(ListA) - set(ListB)

        #print(listC)
        if set(ListA) == set(ListB):
            return connect_val.index(ListB)
    return 999999999

def step2(file_cansee_number):
    file_cansee_number_one_num = {}

    connect_val = create_file_num_list_unique(file_cansee_number)

    connect_val.sort(key=len)

    for key,item in file_cansee_number.items():
        one_file_result = []
        for i in item:
            #print(list(set(i)))
            list_to_num = find_diff_num_list(i,connect_val)
            #print(list_to_num)
            one_file_result.append(list_to_num)
        #かぶりがないようにset にしておく
        file_cansee_number_one_num[key] = list(set(one_file_result))
    
    return file_cansee_number_one_num
```

```python
#拡張版かどうかを調べる
#調べるところが少なくなる
def search_extend(connect_val):
    candidate_small_eater = {}

    for one_key_list in connect_val:
        #print("-"*30)
        #print(one_key_list)
        for ser_one_key_list in connect_val:
            #print("#"*10)
            #print(ser_one_key_list)
            listC = set(one_key_list) - set(ser_one_key_list)
            #print(listC)
            if one_key_list == ser_one_key_list:
                continue
            if len(listC) == 0:
                #print("add")
                add_name = one_key_list
                if str(add_name) in candidate_small_eater:
                    candidate_small_eater[str(add_name)] += [ser_one_key_list]
                else:
                    candidate_small_eater[str(add_name)] = [ser_one_key_list]
    return candidate_small_eater
def step2_1(connect_val):
    candidate_small_eater = search_extend(connect_val)
    return candidate_small_eater
```

```python
import ast
#valuesの中身を追加する　かぶりなし
def only_in_key(candidate_small_eater):
    searched_min_list = []

    for z in list(candidate_small_eater.values()):
        for s in z:
            if not s in searched_min_list:
                searched_min_list.append(s)
    return searched_min_list

def step2_2(candidate_small_eater):
    searched_min_list = only_in_key(candidate_small_eater)

    #print(searched_min_list)
    #print("---------")
    search_list_key =[]

    #keyを調べて、valの中になければ追加する
    for candidate_small in candidate_small_eater.keys():

        output = ast.literal_eval(candidate_small)

        if not output in searched_min_list:
            search_list_key.append(output)
    return search_list_key
```

```python
#変化用に調節する

def step2_3(candidate_small_eater, search_list_key):
    how_to_change_list_to_num = {}

    for get_keysearch_list_key in search_list_key:
        getlist_num = []
        for change_to_num in candidate_small_eater[str(get_keysearch_list_key)]:
            getlist_num.append(find_diff_num_list(change_to_num,connect_val))
            #key_change_to_num = is_find_diff_num_list(get_keysearch_list_key.split("."),connect_val)
            #getlist_num.append(change_to_num)     
        how_to_change_list_to_num[find_diff_num_list(get_keysearch_list_key,connect_val)] = getlist_num
    return how_to_change_list_to_num
```

```python
def is_change(value,how_to_change_list_to_num):
    for keys, items in how_to_change_list_to_num.items():
        #print(f"---{value}----")
        #print(int_list,items)
        if value in items:
            return True , keys
    return False , 99999
#ここで変換をして新しい辞書を作っている

def step2_4(file_cansee_number_one_num,how_to_change_list_to_num):
    trans_file_cansee_number_one_num = file_cansee_number_one_num.copy()

    for keys,values in list(file_cansee_number_one_num.items()):

        new_val = []
        for value in values:
            #変換先があるかどうかを調べる。
            transnum = is_change(value,how_to_change_list_to_num)
            if transnum[0]:

                new_val.append(transnum[1])
                #print(f"{value} to {transnum[1]}")
            else:
                new_val.append(value)

        trans_file_cansee_number_one_num[keys] = new_val

    #ファイルの中身を短縮させた結果
    return trans_file_cansee_number_one_num
```

```python
def get_unique_list(seq):
    seen = []
    return [x for x in seq if x not in seen and not seen.append(x)]


def step3_0_1(trans_file_cansee_number_one_num):
    #valueを取り出し、同じ内容のリストがあれば削除する

    file_tran_to_num = []
    for i in (trans_file_cansee_number_one_num.values()):
        file_tran_to_num.append(sorted(list(set(i))))


    #valuesを取得したあと、かぶってるのを削除あんd、短い順に並び替える
    file_tran_to_num = get_unique_list(file_tran_to_num)
    file_tran_to_num.sort(key=len)
    
    return file_tran_to_num
```

```python
import ast

def step3_0_2(file_tran_to_num):
    #拡張版かどうかを調べる
    candidate_small_eater = search_extend(file_tran_to_num)
    f"""or key,item in (candidate_small_eater.items()):
        print(key)
        print((item))
        print("-"*30)"""


    #上で調べた結果のkeyがvalの中にあるか確認なければ追加
    searched_min_list = only_in_key(candidate_small_eater)
    search_list_key =[]
    #keyを調べて、valの中になければ追加する
    for candidate_small in candidate_small_eater.keys():

        output = ast.literal_eval(candidate_small)

        if not output in searched_min_list:
            search_list_key.append(output)
    return search_list_key,candidate_small_eater

```

```python
def step3_0_3(search_list_key,candidate_small_eater,file_tran_to_num):
    #同要素のとkに抽象化できうるかを調べる

    how_to_change_list_to_num = {}

    for get_keysearch_list_key in search_list_key:
        getlist_num = []
        for change_to_num in candidate_small_eater[str(get_keysearch_list_key)]:
            getlist_num.append(find_diff_num_list(change_to_num,file_tran_to_num))
            #key_change_to_num = is_find_diff_num_list(get_keysearch_list_key.split("."),connect_val)

        how_to_change_list_to_num[find_diff_num_list(get_keysearch_list_key,file_tran_to_num)] = getlist_num

    #'[0, 65]': [16, 38, 46, 70]このとき [0, 65, 16, 38, 46, 70]の中でどれか一つが当てはまれば[0,65]に書き換える

    return how_to_change_list_to_num
```

```python

def step3(file_cansee_number_one_num):

    file_cansee_number_one_num_setted_values = []

    #for i in file_cansee_number_one_num.values():
    for i in file_cansee_number_one_num.values():
        file_cansee_number_one_num_setted_values.append(list(set(i)))
        
    #valueを取り出し、同じ内容のリストがあれば削除する
    file_tran_to_num = []
    for i in (file_cansee_number_one_num_setted_values):
        file_tran_to_num.append(sorted(list(set(i))))


    #valuesを取得したあと、かぶってるのを削除あんd、短い順に並び替える
    file_tran_to_num = get_unique_list(file_tran_to_num)
    file_tran_to_num.sort(key=len)

    return file_cansee_number_one_num_setted_values,file_tran_to_num
```

```python
#共通した組み合わせを探す
def is_find_diff_num_list(ListA,ListB):
 
    #listC = set(ListA) ^ set(ListB)
    #listC = list(listC)
    listC = set(ListA) - set(ListB)

    #print(listC)
    if ListA == ListB:
        return False
    elif len(listC)==0:
        return True
    else:
        return False

def step4(file_cansee_number_one_num_setted_values):
    big_small_dic = {}

    for sear_file_num_list in file_cansee_number_one_num_setted_values:
        #print("-"*20)
        #print(sear_file_num_list)

        inside_list = []
        for in_file_num_list_unique in file_cansee_number_one_num_setted_values:
            if sear_file_num_list == in_file_num_list_unique:
                continue
            if is_find_diff_num_list(sear_file_num_list,in_file_num_list_unique):
                #print("-----inside-----------")
                #print(in_file_num_list_unique)
                inside_list.append(in_file_num_list_unique)
        big_small_dic[str(sear_file_num_list)] = inside_list
    return big_small_dic
```

```python
def step5(big_small_dic):

    adjust_big_small_dic = {}
    count = 1
    for key,item in big_small_dic.items():

        if len(item) > 0:
            add_item = big_small_dic[key]

            for i in list(big_small_dic.values())[count:]:
                for z in i:
                    if z in add_item:
                        add_item.remove(z)
            #print(f"{key} and {len(item)}")

            adjust_big_small_dic[key] = add_item
        count += 1
    return adjust_big_small_dic
```

```python
#拡張場を発見したファイルがどんなふうに拡張する可能性があるのかを可視化する
from graphviz import Digraph
import time
import os
#ファイルを読み込んで、設定されているキーごとに分割をする。それをべうファイルに出力する。

def write_newfile(write_data):
    # ファイルに書き込む
    
    count = len(glob.glob(f"output_templete/" + "*"))
    
    with open(f"output_templete/sample{count}.properties", "w") as file:
        #file.write(f"search_path:{filenames[filenum[0]]} , {filenames[filenum[1]]} \n")
        for text in write_data:
            file.write(text)

def get_num_file_length(num_list):
    write_data = ""
    one_file = []
    for i in num_list:
        one_file.append(connect_val[i])
    print(one_file)
    read_line_length = 0
    for one_user_key_num_data in one_file:
        for one_command_num_data in one_user_key_num_data:
            #print(one_command_num_data)
            with open(f"name_breaker_incom/sample{one_command_num_data}.properties", "r") as file:
                for i in file.readlines():
                    write_data += i
                    print(i[:-1])
                    read_line_length += 1
        print("----------")
    write_newfile(write_data)
    print(f"lenght : {read_line_length}")
    return read_line_length

def bef_creat_see_graph(graph,big_small_dic):
    for key,items in big_small_dic.items():  
        #キーに存在かつ要素の長さが
        if len(items)>0:
            output = ast.literal_eval(key)
            len_key = get_num_file_length(output)
            graph.node(key,f"{key}{len_key}")
            for item in items:
                len_item = get_num_file_length(item)
                graph.node(str(item),f"{item}{len_item}")
                graph.edge(key, str(item))

def step6(adjust_big_small_dic):
    graph = Digraph(format="png")

    bef_creat_see_graph(graph,adjust_big_small_dic)
    print("finish")
    graph.render("image/output")
    # 画像を表示
    graph.view()
```
