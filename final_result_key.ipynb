```python
#上で作成したファイルを調べ、内容が全く同じものがあれば削除する。

#フォルダ削除
import os
#フォルダの整理をする
#上で作成したファイルを調べて同じものがあれば削除する
# フォルダ内のファイルを取得

#はハッシュ値計算
import hashlib

def calculate_hash(file_path):
    hash_object = hashlib.sha256()
    with open(file_path, "rb") as file:
        while True:
            data = file.read(65536)  # ファイルをバッファリング
            if not data:
                break
            hash_object.update(data)
    return hash_object.hexdigest()


#ファイルを削除
def delete_files(folder_path,want_to_delete_name):
    for delete_name in want_to_delete_name:
        file_to_delete = f"{folder_path}{delete_name}"
        if os.path.exists(file_to_delete):
            os.remove(file_to_delete)
            print(f"{file_to_delete} が削除されました。")
        else:
            print(f"{file_to_delete} は存在しません。")

def get_filelist ():
    folder_path = f'name_breaker_incom/'
    file_list = os.listdir(folder_path)

    # 比較するファイルを選択
    file_list = [file_name for file_name in file_list if file_name.endswith('.properties')]
    #print(file_list)
    return folder_path,file_list

def is_delete(filename,want_to_delete_name):
    if filename in want_to_delete_name:
        return True
    else:
        return False
# ファイルを比較
def file_compare(folder_path,file_list):
    want_to_delete_name = []
    delete_file_flag = False
    for file_name in file_list:
        print(f"searcing {file_list.index(file_name)} / {len(file_list)}")
        if is_delete(file_name,want_to_delete_name):
            continue
        
        with open(os.path.join(folder_path, file_name), 'r') as file:
            lines = file.readlines()
        if len(lines) <1:
            #print(file_name)
            want_to_delete_name.append(file_name)
            continue
        # 比較処理を実行
        for other_file_name in file_list[file_list.index(file_name):]:
            if file_name == other_file_name or is_delete(other_file_name,want_to_delete_name):
                continue  # 同じファイルは比較しない

            hash_A = calculate_hash(f"{folder_path}{file_name}")
            hash_B = calculate_hash(f"{folder_path}{other_file_name}")

            if hash_A == hash_B:
                #print("ファイルAとファイルBは同じ内容です。")
                delete_file_flag =True
                #print(f"{file_name} : {other_file_name}")
                want_to_delete_name.append(other_file_name)

        #同じところを調べずに済むようにする
    return delete_file_flag,folder_path,want_to_delete_name

        
```

```python
import os
#ファイルを読み込んで、設定されているキーごとに分割をする。それをべうファイルに出力する。

def write_newfile(write_data, count):
    # ファイルに書き込む
    print("write")
    with open(f"name_breaker_incom/sample{count}.properties", "w") as file:
        #file.write(f"search_path:{filenames[filenum[0]]} , {filenames[filenum[1]]} \n")
        for text in write_data:
            file.write(text)
    count += 1
    
    return count
```

```python
#ユーザ設定のキーを統一する。
def command_to_userkey(block_list):
    result = []
    for one_text in block_list:
        one_command_block = []
        #print(one_text)
        for command in one_text.split('=')[0].split('.'):
            if command.isdecimal():
                one_command_block.append('0')
                continue
            elif command not in commnad_states:
                #print(command)
                one_command_block.append("user_key")
            else:
                one_command_block.append(command)
        result.append(".".join(one_command_block) + "=\n")
    return result
    

def is_same_path(adjusted_block_list,folder_path,file_list):
    for file_name in file_list:
        with open(os.path.join(folder_path, file_name), 'r') as file:
            in_dic_lines = file.readlines()
        
        if adjusted_block_list == in_dic_lines:
            return True
        
    return False
```
