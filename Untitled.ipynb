```python
#使われたコマンドの回数をかぞえて並び変えるを取り出す
def find_max_keys_counts(words_counts):
    #ここで各コマンドの最大値を持ってくる
    max_count_all_files = []
    for words_count in words_counts:
        words_count = sorted(words_count.items(),key=lambda x:x[1],reverse=True)
        words_count = dict((x, y) for x, y in words_count)
        add_graf = {}
        count = 1
        for k,i in words_count.items():
            add_graf[k] = i
            count += 1
            #countが値を超えると新しい辞書型に代入するのをやめる。上位(count)個を取得する。
            if(count > 10):
                break
        max_count_all_files.append(add_graf)
    #print(words_counts)
    return max_count_all_files
#取得したコマンドの文字数を調節する。後どのコマンドが何回使われたのか数える。
def count_words(result):
    words_counts = {}
    
    for search_words in result:
        #文字数が多い時は15文字にする
        if search_words in words_counts:
            words_counts[search_words[:15]] += 1
        else:
            words_counts[search_words[:15]] = 1
    #print(words)
    return words_counts

#コマンドの出現回数を平均にする
import copy
def average_wordcounts(original_all_file_words_counts):
    all_file_words_counts = copy.deepcopy(original_all_file_words_counts)
    for all_file_words_count in all_file_words_counts:
        #print(all_file_words_count)
        word_sum = sum(list(all_file_words_count.values()))
        #print(word_sum)
        for file_words_count in all_file_words_count:
            all_file_words_count[file_words_count] = (all_file_words_count[file_words_count]/word_sum)*100
            #all_file_words_count[words_count] = round(all_file_words_count[words_count]/word_sum,2)
    return all_file_words_counts
```
