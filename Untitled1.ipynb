```python
import hashlib
import os

# フォルダのパスを指定
#folder_path = '//'

# フォルダ内のファイルをリストアップ
file_list = glob.glob("from_github_file_clear/"+"*")
print(len(file_list))
# ファイルのハッシュ値を格納する辞書
hash_dict = {}

removed_filename = []
# フォルダ内のファイルのハッシュ値を計算し、辞書に格納
for file_name in file_list:
    #file_path = os.path.join(folder_path, file_name)
    file_path = file_name
    if os.path.isfile(file_path):
        with open(file_path, 'rb') as file:
            file_hash = hashlib.md5(file.read()).hexdigest()
            if file_hash in hash_dict:
                # 同じハッシュ値を持つファイルがすでに存在する場合、重複ファイルとして処理
                print(f"重複ファイル: {file_path} と {hash_dict[file_hash]}")
                removed_filename.append(file_path)
            else:
                hash_dict[file_hash] = file_path

for removed_file in removed_filename:
    os.remove(removed_file)
    
file_list = glob.glob("from_github_file_clear/"+"*")
print(len(file_list))
print("finish")
```

```python
#プロパティ化
#これを使う前にファイルの中でタブ文字を使っているのを空白に変換する必要がある
import yaml

def yaml_to_properties(yaml_data):
    properties = []

    def traverse(data, path=[]):
        if isinstance(data, dict):
            for key, value in data.items():
                traverse(value, path + [key])
        elif isinstance(data, list):
            for index, value in enumerate(data):
                traverse(value, path + [str(index)])
        else:
            properties.append(".".join(path) + "=" + str(data).replace("\n"," "))

    traverse(yaml_data)
    return "\n".join(properties)

# YAMLファイルの読み込み
diff_all_foldernames = glob.glob("from_github_file_clear/"+"*")
#diff_all_foldernames = ["from_github_history_allcode_clear\\filenum_1_0_0\\2021-06-08T15-40-34Z_config.yml"]
print("yaml to prpperties")
for file in diff_all_foldernames:
    
    try:
        with open(file, "r") as yaml_file:
            yaml_data = yaml.safe_load(yaml_file)
    except:
        print("error")
        continue
    # YAMLデータをプロパティファイル形式に変換
    properties_content = yaml_to_properties(yaml_data)

    # プロパティファイルに保存
    save_file_name = "from_github_file_clear_proper/" + str(file.split("\\")[-1].split(".")[0]) + ".properties"
    with open(save_file_name, "w") as properties_file:
        properties_file.write(properties_content)
    print("ファイルに保存されました: " + str(save_file_name))
```

```python
from janome.tokenizer import Tokenizer
from sklearn.feature_extraction.text import TfidfVectorizer


documents = [
        "publish_preleases.workflows.jobs.2.publish_prerelease.requires.0=setup",

        "workflows.publish_preleases.jobs.2.publish_prerelease.requires.0=Publish to Next channel",

        "jobs.publish_preleases=workflows.2.publish_prerelease.requires.0.Next channel",

        "workflows.requires,publish_preleases=jobs.2.publish_prerelease.0.Publish to Next"
    ]
def use_tf_idf(documents=documents):
    # 形態素解析器を初期化
    tokenizer = Tokenizer()
    
    # 文書をトークン化して単語リストを作成
    tokenized_documents = [' '.join(token.surface for token in tokenizer.tokenize(doc)) for doc in documents]

    # TF-IDFベクトル化器を作成
    vectorizer = TfidfVectorizer()

    # 文書をTF-IDFベクトルに変換
    tfidf_matrix = vectorizer.fit_transform(tokenized_documents)

    # TF-IDFベクトルを取得
    tfidf_vector = tfidf_matrix.toarray()

    # 特徴語（キーワード）を取得
    features = vectorizer.get_feature_names()

    # 結果を表示
    print("特徴語:", features)
    print("TF-IDFベクトル:")
    print(tfidf_vector)
    
    return tfidf_vector
```

```python
import numpy as np

def cos_simmilar(tfidf_vector, num):
    # 2つのベクトルを定義
    simillar_command = []
    vector1 = tfidf_vector[num]
    for i in range(num+1,len(tfidf_vector)):
        vector2 = tfidf_vector[i]

        # ベクトルを正規化
        norm_vector1 = np.linalg.norm(vector1)
        norm_vector2 = np.linalg.norm(vector2)

        if norm_vector1!=0 and norm_vector2 != 0:
            normalized_vector1 = vector1 / norm_vector1
            normalized_vector2 = vector2 / norm_vector2
        else:
            #print("noen")
            continue
        # コサイン類似度を計算
        cosine_similarity = np.dot(normalized_vector1, normalized_vector2)

        if cosine_similarity >= 1:
            #print(items_list_allfile[0][num])
            #print(items_list_allfile[0][num+1])
            continue
        elif cosine_similarity == 0:
            continue
        elif cosine_similarity >= 0.5:# and cosine_similarity <= 0.1:# and False:
            #print(all_setted_commands_join[num])
            simillar_command.append(setted_commands[i])
        # 結果を表示
        #print("コサイン類似度:", cosine_similarity)
    return simillar_command

for z in range(0,len(tfidf_vector)-1):
    simillar_command = cos_simmilar(tfidf_vector,z)
    
    if len(simillar_command) > 0:
        print(f"simmilar sentence \n{setted_commands[z]} \n")
        for i in simillar_command:
            print(i)
        print("#" * 40)
```
